# Aula 5.3 - Construtores e Introdução a Classes

 ### Construtores
 
 **<nome_do_objeto\>= <Classe\> (<argumentos\>)**

```python
# Criando uma classe

# para ter um construtor, precisa sempre ter esse __init__
class Carro:
    def __init__(self, numero_portas, preco): # CONSTRUTOR DA CLASSE
        self.numero_portas = numero_portas
        self.preco = preco
        print('Objeto carro instanciado com sucesso.')

    # metodo get
    def get_numero_portas(self):
        return self.numero_portas

# chamando o construtor
carro_1 = Carro(4,50000)
```

```python
# podemos utilizar os métodos desse objeto

portas_carro_1 = carro_1.get_numero_portas()
print(portas_carro_1)
```

```python
# instanciando outro objeto

carro_2 = Carro(2,80000)
portas_carro_2 = carro_2.get_numero_portas()

print(portas_carro_2)
```

```python
# duas instancias diferentes!

print('Endereço de memória carro_1: ', hex(id(carro_1)))
print('Endereço de memória carro_2: ', hex(id(carro_2)))
```

### Podemos ter mais de um construtor para uma classe!

```python
string = str(6)
print(string)

inteiro = int('6')
print(inteiro)

# construtor recebendo uma string
float_1 = float('6')
print(float_1)

# construtor recebendo um inteiro
float_2 = float(inteiro)
print(float_2)

lista = list('abc')
print(lista)
```

## Métodos acessores e modificadores (get e set)

- **get:** recupera um dado armazenado em um dado objeto

- **set:** altera um dado armazenado em um dado objeto

```python
# importando o módulo array
import array as ar

# inicializando o array
meu_array = ar.array('i', [5,1,3,4,2,2,7,2])

print(meu_array)

# acessando o índice de valor igual a 2
print(meu_array.index(2))

# acessando a quantidade de valores iguais a 2 existentes
print(meu_array.count(2))
```

```python
# criando uma classe Carro
class Carro:
    def __init__(self, numero_portas, preco):
        self.numero_portas = numero_portas
        self.preco = preco
        print('Objeto carro instanciado com sucesso.')

    # método para acessar o valor do nº de portas
    def get_numero_portas(self):
        return self.numero_portas

    # método para atualizar o valor do nº de portas
    def set_numero_portas(self, novo_numero_portas):
        self.numero_portas=novo_numero_portas
        
carro_3 = Carro(4,50000)

# metodo get - não modifica o estado do objeto
print('Nº de portas antes:', carro_3.get_numero_portas())

# metodo set - modifica o estado do objeto
carro_3.set_numero_portas(2)
print('Novo nº de portas:', carro_3.get_numero_portas())
```

#### Criando 2 objetos e igualamos um ao outro, referenciando um mesmo objeto

```python
class Carro:
    def __init__(self, numero_portas, preco):
        self.numero_portas = numero_portas
        self.preco = preco
        print('Objeto carro instanciado com sucesso.')

    # método para acessar o valor do nº de portas
    def get_numero_portas(self):
        return self.numero_portas

    # método para atualizar o valor do nº de portas
    def set_numero_portas(self, novo_numero_portas):
        self.numero_portas=novo_numero_portas

# instanciando objeto carro_3
carro_3 = Carro(4,50000)

# metodo get - não modifica o estado do objeto
print('Nº de portas antes:', carro_3.get_numero_portas())

# metodo set - modifica o estado do objeto
carro_3.set_numero_portas(2)

# cria a referência ao objeto carro
carro_4 = carro_3
print('Nº de portas carro_4:', carro_4.get_numero_portas())

# modificando o numero de portas do carro_4
carro_4.set_numero_portas(3)
print('Novo nº de portas do carro_4:', carro_4.get_numero_portas())

# numero de portas do carro_3
print('Nº de portas carro_3:', carro_3.get_numero_portas())
```

Primeiro tínhamos carro3 com 4 portas. Depois, utilizamos o método set_numero_portas() e alteramos para 2 portas.

Após isso, criamos uma referência ao objeto carro, com carro_4 = carro_3.

Então, modificamos o numero de portas do carro4, e automaticamente, por estarem referenciando um mesmo objeto (numa mesma posição de memória), o carro3 tem seu número de portas atualizado também.

Em resumo, como carro3 e carro4 estão apontando para a mesma posição de memória, quando modificamos (com set) o valor de um, o valor do outro se atualiza também.

#### Classes imutáveis

```python
# classes que não possuem métodos mutáveis: int, float, string
string = 'Estudando Python '
print(hex(id(string)))

# criando um novo objeto, numa nova posição de memória
string = string + 'no IGTI'
print(hex(id(string)))

# por isso é possível a concatenação dos valores!
print(string)
```

# Aula 5.4 - Arquivos XML

Extensible Markup Language

```python
# importar módulo para leitura de xml
from xml.dom import minidom

nome_arquivo_xml = 'arquivo_xml.xml' # nome do arquivo a ser importado
documento_xml = minidom.parse(nome_arquivo_xml) # instancia o objeto
```

```python
documento_xml
```

```python
# nós presentes no documento
nos = documento_xml.documentElement
print(nos.tagName)
```

```python
# extraíndo elementos de um arquivo xml
# cada elemento em um arquivo xml tem um nome
lista_de_compras = documento_xml.getElementsByTagName('item')
type(lista_de_compras)
```

```python
lista_de_compras[0]
```

```python
# acessando os atributos de um arquivo
for item in lista_de_compras:
    print(item.attributes['name'].value)
```

```python
# acessando os elementos (valores) de um arquivo
for item in lista_de_compras:
    print(item.firstChild.data) # pegando os elementos filhos
```
