Servidor Web:







### O que é Django?
- Web framework escrito em Python;
- Software livre BSD-Berkeley Software Distribution;
- Semelhante ao ASP e ao JSP;
- Gera páginas estáticas;
- Não é um SPA (Single Page App).

### Premissas
- Reusabilidade e Plugabilidade (apps);
- Alto desacoplamento;
- Rápido desenvolvimento;
- Don't repeat yourself (DRY).

### Modelo MVC
- **Model:** modelo de dados;
- **View:** visualização;
- **Controller:** controla as ações entre V e M.

### Modelo MTV = MVC sem estado
- **M:** Modelo;
- **T:** Template;
- **V:** Controller.

### Alguns conceitos
- **Project:** projeto web completo;
- **App:** módulos do projeto:
    - Apps próprios;
    - Apps de terceiros.
    
### Templates:
- páginas HTML que permite código; 
- podem ser extremamente reutilizáveis; 
- recebem informações vindas das views.
