Servidor Web:







### O que  Django?
- Web framework escrito em Python;
- Software livre BSD-Berkeley Software Distribution;
- Semelhante ao ASP e ao JSP;
- Gera pginas estticas;
- No  um SPA (Single Page App).

### Premissas
- Reusabilidade e Plugabilidade (apps);
- Alto desacoplamento;
- Rpido desenvolvimento;
- Don't repeat yourself (DRY).

### Modelo MVC
- **Model:** modelo de dados;
- **View:** visualizao;
- **Controller:** controla as aes entre V e M.

### Modelo MTV = MVC sem estado
- **M:** Modelo;
- **T:** Template;
- **V:** Controller.

### Alguns conceitos
- **Project:** projeto web completo;
- **App:** mdulos do projeto:
    - Apps prprios;
    - Apps de terceiros.
    
### Templates:
- pginas HTML que permite cdigo; 
- podem ser extremamente reutilizveis; 
- recebem informaes vindas das views.
